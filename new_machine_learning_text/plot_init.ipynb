```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f3(x0, x1):
    r = 2*x0 **2 +x1 **2
    ans = r*np.exp(-r)
    
    return ans
```

```python
xn = 9
x0 = np.linspace(-2,2, xn)
x1 = np.linspace(-2, 2, xn)
```

```python
y = np.zeros((len(x0), len(x1)))
```

```python
y
```

```python
for i0 in range(xn):
    for i1 in range(xn):
        y[i0,i1] = f3(x0[i0], x1[i1])
```

```python
print(x0)
```

```python
y
```

```python
np.round(y, 1)
```

```python
plt.figure(figsize=(3.5, 3))
plt.gray()
plt.pcolor(y)
plt.colorbar()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
xx0, xx1 = np.meshgrid(x0, x1)
```

```python
plt.figure(figsize=(5, 3.5))
ax = plt.subplot(1,1,1, projection='3d')
ax.plot_surface(xx1, xx0, y, rstride=1, cstride=1, alpha=0.3, 
                color='blue',edgecolor='black')
ax.set_zticks((0, 0.2))
ax.view_init(75, -95)
plt.show()
```
