```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
import numpy as np
from numpy import random
```

```python
random.seed(0)
```

```python
x = np.random.randn(30)
```

```python
y = np.sin(x) + np.random.randn(30)
```

```python
plt.figure(figsize=(20,6))
plt.plot(x, y, 'o')
plt.title("Sample plot")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

```python
np.random.seed(0)
np_data_X = np.arange(1000)
```

```python
np_random_data_Y = np.random.randn(1000).cumsum()
```

```python
plt.figure(figsize=(20,6))
plt.plot(np_data_X, np_random_data_Y, label='Label1')
plt.grid(True)
```

```python
plt.figure(figsize=(20,6))
plt.subplot(2,1,1)

x = np.linspace(-10,10,100)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
y = np.linspace(-10,10,100)
plt.plot(y, np.sin(2*y))

plt.grid(True)
```

```python
def my_func(x):
    return x ** 2 + 2*x +1

x = np.arange(-10,10)

plt.figure(figsize=(20,6))
plt.plot(x, my_func(x))

plt.grid(True)
```

```python
np.random.seed(0)

plt.figure(figsize=(20,6))

plt.hist(np.random.randn(10**5)*10+50, bins=60, range=(20,80))
```
